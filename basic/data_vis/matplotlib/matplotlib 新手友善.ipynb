## 字體配置

```python
# 坐標軸標題，圖表標題
plt.plot(X,Y, color="lime", linestyle="-", linewidth=2, \
         marker="*", markerfacecolor= "red", markeredgecolor="blue" )

ax = plt.gca() # 要產生座標軸要用的套件

ax.set_title("sin(x) function") # 圖的標題
ax.set_xlabel("time (sec)") # X軸名稱
ax.set_ylabel("Intensity (W)") # Y軸名稱
```

```python
# 坐標軸標題，圖表標題，字形配置
plt.plot(X,Y, color="lime", linestyle="-", linewidth=2, \
         marker="*", markerfacecolor= "red", markeredgecolor="blue" )

# 把軸取出來
ax = plt.gca()

# 粗斜體
ax.set_title("sin(x) function", fontname = "Arial", fontsize=20, weight="bold", style="italic")
ax.set_xlabel("Time (s)", fontname = "Arial", fontsize=13)
ax.set_ylabel("Intensity (W)", fontname = "Arial", fontsize=13)
```

## 坐標軸刻度配置

```python
# 坐標軸標題，圖表標題，字形配置
plt.plot(X,Y, color="lime", linestyle="-", linewidth=2, \
         marker="*", markerfacecolor= "red", markeredgecolor="blue" )

ax = plt.gca()

ax.set_title("sin(x) function", fontname = "Arial", fontsize=20)
ax.set_xlabel("Time (s)", fontname = "Arial", fontsize=13)
ax.set_ylabel("Intensity (W)", fontname = "Arial", fontsize=13)

ax.set_xticks([0, 2.5, 5, 7.5, 10, 12.5]) # 改X軸的刻度
```

```python
# 坐標軸標題，圖表標題，字形配置
plt.plot(X,Y, color="lime", linestyle="-", linewidth=2, \
         marker="*", markerfacecolor= "red", markeredgecolor="blue" )

ax = plt.gca()

ax.set_title("sin(x) function", fontname = "Arial", fontsize=20)
ax.set_xlabel("Time (s)", fontname = "Arial", fontsize=13)
ax.set_ylabel("Intensity (W)", fontname = "Arial", fontsize=13)

# 換標籤座標軸
_ = ax.set_xticklabels(["A", "B", "C", "D", "E", "F", "G", "H"])
```

```python
# 坐標軸標題，圖表標題，字形配置
plt.plot(X,Y, color="lime", linestyle="-", linewidth=2, \
         marker="*", markerfacecolor= "red", markeredgecolor="blue" )

ax = plt.gca()

ax.set_title("sin(x) function", fontname = "Arial", fontsize=20)
ax.set_xlabel("Time (s)", fontname = "Arial", fontsize=13)
ax.set_ylabel("Intensity (W)", fontname = "Arial", fontsize=13)

ax.set_xticklabels(["A", "B", "C", "D", "E", "F", "G", "H"])
# 座標軸刻度線調整
ax.tick_params(axis="both", direction = "in", color="orange", length=10, width=3)
```

## 多線圖

```python
plt.plot(X,Y)
plt.plot(X+3,Y)
plt.plot(X,Y+2)
plt.plot(X+3,Y+2)
```

### 設定標籤

```python
# 坐標軸標題，圖表標題，字形配置
plt.plot(X,Y, linestyle="solid", linewidth=5, label="noraml")
plt.plot(X+3,Y, linestyle="--", linewidth=5, label="shift 3 time period")

ax = plt.gca()

ax.set_title("sin(x) function")
ax.set_xlabel("Time (s)")
ax.set_ylabel("Intensity (W)")

plt.legend() # 圖示
```

```python
# 坐標軸標題，圖表標題，字形配置
plt.plot(X,Y, linestyle="solid", linewidth=5, label="noraml")
plt.plot(X+3,Y, linestyle="--", linewidth=5, label="shift 3 time period")

ax = plt.gca()

ax.set_title("sin(x) function")
ax.set_xlabel("Time (s)")
ax.set_ylabel("Intensity (W)")

plt.legend(loc = "lower left")
```

```python
# zorder 圖Z層的順序
plt.plot(X,Y, linestyle="solid", linewidth=5, label="noraml", zorder=4) #藍
plt.plot(X+3,Y, linestyle="--", linewidth=5, label="shift 3 time period", zorder=2) #橘
plt.plot(X+6,Y, linestyle="--", linewidth=5, label="shift 3 time period", zorder=0) #綠

ax = plt.gca()

ax.set_title("sin(x) function")
ax.set_xlabel("Time (s)")
ax.set_ylabel("Intensity (W)")

plt.legend(loc = "lower left")

```

## 子圖

```python
# 開畫布 fig
fig, ax = plt.subplots(2,1) # (2,1) = row*2 col*1
```

```python
# 開畫布 2x1
fig, ax = plt.subplots(2,1)

ax[0].plot(X,Y) # 上面的
ax[1].plot(X,Y, 'r') # 下面的
```

```python
# 開畫布 3x2
fig, ax = plt.subplots(3,2)
ax[2,1].plot(X,Y)

ax[1,0].plot(X+3,Y) # X軸座標有改變
```

```python
fig, ax = plt.subplots(2,1)

ax[0].plot(X,Y)
ax[1].plot(X+3,Y, 'r')

# 配置座標軸範圍
ax[0].set_xlim([0,10])
ax[1].set_xlim([0,10])

```

```python
fig, ax = plt.subplots(2,1)

x = np.arange(1,10)
y = np.exp(x)

ax[0].plot(x,y)
ax[1].plot(x,y, 'r')

# log座標變換
ax[1].set_yscale('log') # 對數軸
```

### 雙y軸

```python
x = np.arange(1,10)
y = np.exp(x)

ax = plt.gca()

plt.plot(x,y)
# plt.plot(x,y*100)

# 共用x軸
ax2 = ax.twinx()

ax2.plot(x+3,y*10,"r")
```

### 雙x軸

```python
x = np.arange(1,10)
y = np.exp(x)

ax = plt.gca()

plt.plot(x,y)

# 共用y軸
ax2 = ax.twiny()

ax2.plot(x+3,y/10,"r")
```

## 輸出圖表

```python
# 改變畫布大小
fig = plt.figure(figsize=(8,4))

# 建立假數據
x = np.arange(1,10)
y = np.exp(x)

# 取出軸資訊
ax = plt.gca()

# 畫圖
plt.plot(x,y)

# 共用y軸
ax2 = ax.twiny()

ax2.plot(x+3,y/10,"r")

# 存儲圖片, 設定解析度
plt.savefig('./view.png', dpi=500) # dpi設定解析度
```
